```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header=1)
df.head()
```

```python
df.info()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122, 'Region'] = 0
df.loc[122:, 'Region'] = 1
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
df.info()
```

```python
df = df.dropna().reset_index(drop= True)
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] .astype(int)
df.info()
```

```python
objects = [features for features in df.columns if df[features].dtype == 'O'] ### O means object type
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```
