```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header=1)
df.head()
```

```python
df.info()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122, 'Region'] = 0
df.loc[122:, 'Region'] = 1
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
df.info()
```

```python
df = df.dropna().reset_index(drop= True)
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] .astype(int)
df.info()
```

```python
objects = [features for features in df.columns if df[features].dtype == 'O'] ### O means object type
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv('Algerian_forest_fires_dataset_CLEANED.csv')
```

```python
df = df.drop(['day', 'month', 'year'], axis=1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
df['Classes'] = np.where(df['Classes'].str.contains('not fire'), 0 , 1)
df['Classes'].value_counts()
```

```python
plt.style.use('seaborn')
df.hist(bins= 50, figsize = (20, 15))
plt.show()
```

```python
percentages =  df['Classes'].value_counts(normalize = True) * 100
class_labels = ['Fire', 'Not Fire']
plt.figure(figsize=(12, 7))
plt.pie(percentages, labels= class_labels, autopct = "%1.1f%%")
plt.title('Pie chart of classes')
plt.show()
```
